```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
df=pd.read_excel("https://archive.ics.uci.edu/ml/machine-learning-databases/00352/Online%20Retail.xlsx")
df
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates()
```

```python
df['Description']=df['Description'].str.strip()
```

```python

sns.histplot(df['Country'])
plt.show()
```

```python
df['Country'].unique()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
basket=df.loc[df['Country']=='France'].groupby(['InvoiceNo','Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo')
basket
```

```python
def encode(x):
  if x>0:
    return 1
  if x<=0:
      return 0
encoded_basket=basket.applymap(encode)
encoded_basket
```

```python
apr=apriori(encoded_basket,min_support=0.07,use_colnames=True)
apr
```

```python
rules=association_rules(apr)
rules=rules.sort_values(['confidence'],ascending=False)
rules
```

```python
rules[(rules['lift']>=3)&(rules['confidence']>=0.3)]
```

```python
df1=df.loc[df['Country']=='Germany'].groupby(['InvoiceNo','Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo')
df1
```

```python
def encode(x):
  if x>0:
    return 1
  if x<=0:
      return 0
encoded_df1=df1.applymap(encode)
encoded_df1
```

```python
apri=apriori(encoded_df1,min_support=0.07,use_colnames=True)
apri
```

```python
rule=association_rules(apri,metric="lift",min_threshold=1)

rule
```

```python
def get_recommendations(product,rules,num_recommendations =5):
  recommendations=[]
  for idx,row in rule.iterrows():
    if product in row['antecedents']:
      recommended_item=next(iter(row['consequents']))
      recommendations.append((recommended_item,row['lift']))
  recommendations.sort(key=lambda x:x[1],reverse=True)
  return recommendations[:num_recommendations]
```

```python
product_to_recommend='REGENCY CAKESTAND 3 TIER'
recommendations=get_recommendations(product_to_recommend,rule)
print(f"Recommendations for'{product_to_recommend}':")
for item,lift in recommendations:
  print(f"{item}(lift:{lift:.2f})")
```
